```python
# import 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
# Read the data
# This data you can find here: https://www.kaggle.com/c/home-data-for-ml-course/data

X_full = pd.read_csv('train.csv', index_col='Id')
X_test_full = pd.read_csv('test.csv', index_col='Id')

# SalePrice is the target, if there is no target eliminate row associated with it
X_full.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = X_full.SalePrice
X = X_full.copy()
X.drop(['SalePrice'], axis=1, inplace=True)



X_test = X_test_full.copy()


# Now we have X and y the target separate! 
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
print(X.shape)
X.head(1)
```

```python
y.head()
```

# Baseline model (deleting every column that has NULL)

```python
X_drop = X.copy()
X_drop.dropna(axis=1,inplace=True)
```

```python
X_drop.shape
# now we don't have null so the second parameter is irrelevant
X_drop = pd.get_dummies(X_drop,dummy_na=False)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_drop, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
model = RandomForestRegressor(n_estimators=50, random_state=0)
model.fit(X_train,y_train)
predictions = model.predict(X_valid)
mae_base = mean_absolute_error(y_valid, predictions)
print('MAE_0: ', mae_base )
```

```python
missing_val_count_by_column = (X.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
# notice these are just the number columns we deal with nan with pd.get_dummies() above
```

Information about columns
+ LotFrontage: Linear feet of street connected to property
+ GarageYrBlt: Year garage was built

probably the reason for missing values:
+ Negligent amount of feet connected to the property

Notice that in order to fill correctly these null values we need to know a little more about the business and do a more in depth analysis, this is behond of the scope of this Jupyter's Notebook.
So will eliminate the columns with too much missing values.

```python
# If there are columns that have and immense quantities of null 
# I will eliminate them that has more than 10% of missing values 


percentage_permited = 10
columns_to_delete = list(missing_val_count_by_column[missing_val_count_by_column > X.shape[0]*(percentage_permited/100)].index)
X.drop(columns=columns_to_delete, inplace=True, axis=1)
X.shape
```

```python
# six less columns these were the columns deleted
columns_to_delete
```

## Numerical Columns
We are dealing with missing variables, but depending of the type of variables we will treat them in a different manner. We will use here SimpleImputer. This has different parameters to choose and we will use that to treat differently each of our missing data (depending if they are categorical or numerical ones).

```python
# all columns that have missing values
# remember we eliminate some columns so now we have less of them
missing_val_count_by_column = (X.isnull().sum())
columns_need_treatment = list(missing_val_count_by_column[missing_val_count_by_column > 0].index)
```

```python
X[columns_need_treatment].columns
```

```python
X[columns_need_treatment].dtypes.unique()
```

```python
numerical_col = [col for col in columns_need_treatment if str(X[col].dtypes)=='float64' ]
numerical_col
```

before applying this imputer functions we are gonna copy the X to a new X_copy because we will want to see if imputer values is better than just actually droping these columns altogether.
Notice that if:

+ there is a consistency in the missing of variable imputer is not the best. It is best to use  fillna like this: df.fillna(0).
+ We can add another column that show if there is a missing value, and we can analyse those situation separetely (we could use cross entropy to see if there is other variable that is influence by this one)

```python
X_copy = X.copy()
```

```python
X_copy.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_copy, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
# Library to solve the missing values
from sklearn.impute import SimpleImputer 
# put mean when we have a null
my_num_imputer = SimpleImputer(strategy='mean')
X_train[numerical_col] = my_num_imputer.fit_transform(X_train[numerical_col])
X_valid[numerical_col] = my_num_imputer.transform(X_valid[numerical_col])
X_train.head()
```

```python
missing_values = X_train.isnull().sum()
missing_values[missing_values>0]
```

```python
# sin valores nulos!!
```

```python
model = RandomForestRegressor(n_estimators=50, random_state=0)
model.fit(X_train,y_train)
predictions = model.predict(X_valid)
mae_1 = mean_absolute_error(y_valid, predictions)
print(f"MAE_1: {mae_1} vs MAE_0 {mae_base}")
```
